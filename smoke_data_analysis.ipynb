```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tobacco_data = pd.read_csv("tobacco.csv")
tobacco_data
```

```python
tobacco_data.isnull().sum()
```

```python
sns.heatmap(tobacco_data.isnull())
plt.show()
```

```python
tobacco_data["Location 1"].mode()[0]
```

```python
tobacco_data.fillna(tobacco_data["Location 1"].mode()[0] , inplace =True)
```

```python
tobacco_data.isnull().sum()
```

```python
tobacco_data["Year"].value_counts()
```

```python
tobacco_data["State"].value_counts()
```

```python
plt.figure(figsize = (15,5))
sns.scatterplot(x = "State" , y = "Year" , data = tobacco_data)
plt.xticks(rotation= 90)
plt.show()
```

```python
tobacco_data.info()
```

```python
tobacco_data.describe()
```

```python
tobacco_data["Smoke everyday_float"]= tobacco_data["Smoke everyday"].str.replace("%" ,"").astype(float)
```

```python
tobacco_data["Smoke some days_float"]= tobacco_data["Smoke some days"].str.replace("%" ,"").astype(float)
```

```python
tobacco_data["Former smoker_float"]= tobacco_data["Former smoker"].str.replace("%" ,"").astype(float)
```

```python
tobacco_data["Never smoked_float"]= tobacco_data["Never smoked"].str.replace("%" ,"").astype(float)
```

```python
tobacco_data
```

```python
plt.figure(figsize = (15,5))
sns.stripplot(x= "State" , y = "Smoke everyday_float" , data = tobacco_data ,hue ="State",palette = "viridis_r")
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.stripplot(x= "State" , y = "Smoke some days_float" , data = tobacco_data ,hue ="State",palette ="gist_earth")
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.stripplot(x= "State" , y = "Former smoker_float" , data = tobacco_data ,hue ="State",palette ="gist_earth")
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.stripplot(x= "State" , y = "Former smoker_float" , data = tobacco_data ,hue ="State", palette ="Dark2")
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.pairplot(tobacco_data,hue = "State" , kind ="hist",diag_kind = "kde")

plt.show()
```
